```python
import pandas as pd
import numpy as np
import math
```

```python
data = pd.read_csv('DATA.csv')
df = pd.DataFrame(data,columns=['F1','F2','F3','F4','F5'])
df
```

```python
p1 = len(df)
p2 = len(df)
p3 = len(df.loc[0])
arr = []
hasil = []
for i in range(p1):
    arr1 = []
    hasil1 = []
    for j in range(p2):
        arr2 = []
        hasil2= []
        for k in range(p3):
            z = df.loc[i][k]-df.loc[j][k]
            m = 0
            if (z <= 0):
                m = 0
            else:
                m = 1
            hasil2.append(z)
            arr2.append(m)
        arr1.append(arr2)
        hasil1.append(hasil2)
    arr.append(arr1)
    hasil.append(hasil1)
```

```python
hasil1 = pd.DataFrame(arr)
hasil2 = pd.DataFrame(hasil)
```

```python
akhir = []
for i in range(len(hasil)):
    akhir1 = []
    for j in range(len(arr[i])):
        bla = sum(arr[i][j])/len(arr[i][j])
        akhir1.append(bla)
    akhir.append(akhir1)
akhir
```

```python
tmp = []
for i in range(len(akhir)):
    tmp1 = []
    for j in range(len(akhir[i])):
        tmp1.append(akhir[j][i])
    tmp.append(tmp1)
ip= pd.DataFrame(data=tmp, columns = ['A','B','C','D'], index = ['A', 'B', 'C', 'D'])
ip
```

```python
def lflow():
    lf=[]
    for i in range(len(akhir)):
        jwb = 1/(len(akhir) - 1) * sum(akhir[i])
        lf.append(jwb)
#     lff=pd.DataFrame(data=lf,columns=['leafing flow'])
    return lf

def efl():
    ef = []
    for i in range(len(tmp)):
        jwb = 1/(len(tmp) - 1) * sum(tmp[i])
        ef.append(jwb)
    
#     ent=pd.DataFrame(ef,columns=['Entr Flow'])
    return ef

def nef():
    lf=lflow()
    ef=efl()
    nf = []    
    for i in range(len(ef)):
        jwb = lf[i]-ef[i]
        nf.append(jwb)
#         j = i + 1
#         if (i < len(ef) - 1):
#             jwb = lf[i] - ef[i+1]
#         else:
#             jwb = lf[i] - ef[0]
#         nf.append(jwb)
        nflow=pd.DataFrame(data=nf,columns=['net flow'])
    return nflow
entrflow= pd.DataFrame(data=efl(),columns=['Entry Flow'])    
leavingflow= pd.DataFrame(data=lflow(),columns=['leaving flow'])
```

```python
nef()
```

```python
rangking=pd.concat([entrflow,leavingflow,nef()],axis=1)
rangking
```
